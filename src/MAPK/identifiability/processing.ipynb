```python
fluxes = open('hornberg_2005.jl', 'r')
flux_list = []
for line in fluxes:
    tmp = line.strip('\n').split(' = ')
    tmp[0] = int(tmp[0].strip('dy_dt_'))
    # print(tmp[0])
    flux_list.append(tmp)

flux_list = sorted(flux_list, key=lambda x: x[0])
for item in flux_list:
    print('x' + str(item[0]+1) + "'(t) = " + item[1])
```
